```python
# load libraries 

import pandas as pd
```

```python
# Mean can get affected by outliers and that will skew the distribution of data 
# Median is robust against outliers, so it won't get affected by outliers 

df['Review Rating'] =df.groupby( 'Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# The null values are get replaced using median 
df.isna().sum()
```

```python
# first the columns names are converted to lower case

df.columns = df.columns.str.lower()

# then the blank spaces are replaced with underscore 

df.columns = df.columns.str.replace(' ', '_')
```

```python
# then renamed the 'purchase_amount_(usd)' column 

df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# created a column to categorize the customers according to their age 

labels = ['Young-Adult', 'Adult', 'Middle-aged', 'Senior'] 

df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']]
```

```python
# Created frequently purchased days column 

frequency_mapping ={
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30, 
    'Quarterly' : 90,
    'Bi-Weekly': 14, 
    'Annually' : 365, 
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_days']]
```

```python
df[['discount_applied', 'promo_code_used']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Deleted the unnecessary column 

df = df.drop('promo_code_used', axis=1)
```
